```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la señal x(t) = 7 e^{-6 t} u(t)
def x(t):
    return 7*np.exp(-6*t) * (t >= 0)

# Vector de tiempo
t = np.linspace(-1, 2, 1000)   # desde -1 hasta 2 segundos

# Evaluar la señal
xt = x(t)

# Graficar
plt.figure()
plt.plot(t, xt)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("x(t) = 7 e^{-6 t} u(t)")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)
plt.show()
```

```python
# Intervalo para aproximar la integral (0 a T_grande)
T_grande = 5      # puedes aumentar a 10 si quieres más precisión
N = 100000        # número de puntos para la integral

t_int = np.linspace(0, T_grande, N)
dt = t_int[1] - t_int[0]

# x(t)^2
xt2 = x(t_int)**2

# Integral numérica (regla del rectángulo)
E_num = np.sum(xt2) * dt

print("Energía numérica aproximada:", E_num)
print("Energía teórica (49/12):", 49/12)
```

```python
# Aproximar potencia promedio: P = lim_{T->inf} 1/(2T) ∫_{-T}^{T} |x(t)|^2 dt
T = 5
N = 100000
t_p = np.linspace(-T, T, N)
dt_p = t_p[1] - t_p[0]

P_num = (1/(2*T)) * np.sum(x(t_p)**2) * dt_p
print("Potencia promedio aproximada:", P_num)
```
